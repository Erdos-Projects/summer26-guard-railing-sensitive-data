```python
import pandas as pd
import numpy as np
import csv

imdb = pd.read_csv('imdb_data.csv')

plankton = imdb[imdb['user'].str.contains('plankton', na=False)]

movies = plankton['title'].unique()

movie_titles = pd.read_csv('../data/netflix/movie_titles.csv')

plankton['netflix_id'] = plankton['title'].map(lambda x: int(movie_titles[movie_titles['title'] == x]['id'].values[0]) if x in movie_titles['title'].values else 0) 
plankton = plankton[plankton['netflix_id'] != 0]

movie_ids = movie_titles[movie_titles['title'].isin(movies)][['id']]

plankton['rating_low'] = plankton['rating'] // 2
plankton['rating_high'] = plankton['rating'] // 2 + ((plankton['rating'] % 2) != 0)

df = plankton[['netflix_id', 'rating_low', 'rating_high']]

print(len(movie_ids))
```

```python


input_files = [
    '../data/netflix/combined_data_1.txt',
    '../data/netflix/combined_data_2.txt',
    '../data/netflix/combined_data_3.txt',
    '../data/netflix/combined_data_4.txt'
]

output_file = "netflix_plankton.csv"

with open(output_file, "w", newline="") as out:
    writer = csv.writer(out)
    writer.writerow(["userId", "movieId", "rating"])

    for file in input_files:
        print(f"Processing {file}...")

        with open(file, "r") as f:
            movie_id = None

            for line in f:
                line = line.strip()

                # Movie ID line (e.g., "1:")
                if line.endswith(":"):
                    movie_id = int(line.replace(":", ""))
                else:
                    # userId,rating,date
                    user_id, rating, _ = line.split(",")
                    if movie_id in df['netflix_id'].values:
                        writer.writerow([user_id, movie_id, rating])

print("Done! Saved to netflix_plankton.csv")
```
