- Interesting paper that de-ANONed users in Netflix dataset. Could use this dataset as baseline for our de-ANON
Full Abstract: *We present a new class of statistical de- anonymization attacks against high-dimensional micro-data, such as individual preferences, recommendations, transaction records and so on. Our techniques are robust to perturbation in the data and tolerate some mistakes in the adversary's background knowledge. We apply our de-anonymization methodology to the Netflix Prize dataset, which contains anonymous movie ratings of 500,000 subscribers of Netflix, the world's largest online movie rental service. We demonstrate that an adversary who knows only a little bit about an individual subscriber can easily identify this subscriber's record in the dataset. Using the Internet Movie Database as the source of background knowledge, we successfully identified the Netflix records of known users, uncovering their apparent political preferences and other potentially sensitive information.* LINK: https://arxiv.org/pdf/cs/0610105

- Statistical physics-inspired method that can re-idenity 95% of users out of 1.5 million using geospatial data. Would be interesting to do the same thing but instead of cell phone tower data, browser data. LINK:https://www.erdosinstitute.org/programs/asynchronous/alumni-club-projects-and-ventures

- In 2019, two researchers at the University of Zurich, analyzed cases of the Federal Supreme Court ofSwitzerland to assess which pharmaceutical companies and which medical drugs were involved in legal actions against the Federal Office of Public Health (FOPH) regarding pricing decisions of medical drugs. In general, involved private parties (such as pharmaceutical companies) and information that would reveal the private party (for example, drug names) are anonymized in Swiss judgments. The researchers were able to re-identify 84% of the relevant anonymized cases of the Federal Supreme Court of Switzerland by linking information from publicly accessible databases. LINK: https://jusletter.weblaw.ch/en/juslissues/2019/990/re-identifikation-vo_21cb82c096.html__ONCE&login=false

- In 1997, Latanya Sweeney found from a study of Census records that up to 87 percent of the U.S. population can be identified using a combination of their 5-digit zip code, gender, and date of birth. LINK:  https://dataprivacylab.org/projects/identifiability/paper1.pdf

### Potential Project Ideas

A. We could take a known but nontrivial ML project (e.g. classifying cancer as benign or malignant, insurance fraud detection, etc.), run our anonymizer, and see how well the model still performs while testing the state-of-art (or our own) de-anonymizers on the anonymized data. We would aim for at least 90% accuracy to model with original data and at most 20% of entries re-identitied (can use LLM as baseline de-ANON and ANON). Note: based on what I've read so far, it seems like LLMs could be the **strongest** anonymizers and de-anonymizers (especially for text).

B. Some other narrower scopes we could focus on:
> 1. Big data approach: Traditional anonymization methods (k-anonymity, l-diversity, t-closeness) break down as they lack the scalability to handle big data's volume, velocity, and variety. This leads to both privacy breaches and data quality loss. A potential tree-based (Silva et al.) anonymization algorithm could work.

> 2. Same as 2 but with emphasis on streaming (aka online) data. A cluster-based (Andrew et al.) anonymization algorithm is suggested.

> 3. Same as 2 but with sparse data. For example, we could anonymize the Adult dataset of the 1997 paper (Latanya Sweeney) and run her de-anonymizer, as well as others, to show that they don't work on our ANON.

> 4. Same as 2 but with unstructred data (e.g. graph)
