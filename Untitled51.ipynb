```python
import numpy as np
import pandas as pd

from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

np.random.seed(0)

# Load Adult Census
adult = fetch_openml("adult", version=2, as_frame=True)
df = adult.frame.dropna().copy()

# target/sensitive value
target = "class"

# quasi-identifiers
Q_num = ["age", "capital-gain", "capital-loss", "hours-per-week"]
Q_cat = ["sex", "race", "education", "marital-status", "occupation", "native-country"]

df = df[Q_num + Q_cat + [target]].copy()
df[target] = df[target].astype(str)

print(df.shape)
df.head()
```

```python
def anonymize(df, num_noise=0.10, cat_flip_prob=0.05):
    out = df.copy()

    # numerical multiplicative noise
    for col in Q_num:
        x = out[col].astype(float).values
        noise = np.random.normal(loc=0, scale=num_noise, size=len(out))
        out[col] = np.maximum(0, x * (1 + noise))

    # categorical random perturbation
    for col in Q_cat:
        vals = out[col].astype(str).values
        categories = np.array(sorted(out[col].astype(str).unique()))
        flip = np.random.rand(len(out)) < cat_flip_prob
        vals[flip] = np.random.choice(categories, size=flip.sum())
        out[col] = vals

    return out

anon = anonymize(df, num_noise=0.10, cat_flip_prob=0.05)
anon.head()
```

```python
def random_k_for_row(true_row, released, age_eps=3, hours_eps=5, income_like_eps=0.20):
    mask = np.ones(len(released), dtype=bool)

    # numerical approximate matching
    mask &= np.abs(released["age"] - true_row["age"]) <= age_eps
    mask &= np.abs(released["hours-per-week"] - true_row["hours-per-week"]) <= hours_eps

    for col in ["capital-gain", "capital-loss"]:
        x = true_row[col]
        tol = max(1000, income_like_eps * max(1, x))
        mask &= np.abs(released[col] - x) <= tol

    # categorical exact matching
    for col in ["sex", "race", "education"]:
        mask &= released[col].astype(str).values == str(true_row[col])

    return mask.sum()

sample = df.sample(1000, random_state=1)

Ks = []
for _, row in sample.iterrows():
    Ks.append(random_k_for_row(row, anon))

Ks = np.array(Ks)

print("Mean K:", Ks.mean())
print("Median K:", np.median(Ks))
print("Fraction uniquely identifiable K=1:", np.mean(Ks == 1))
print("Fraction with K<5:", np.mean(Ks < 5))
print("Worst-case K:", Ks.min())
```

```python
X = anon[Q_num + Q_cat]
y = df[target]   # original sensitive value

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0, stratify=y
)

preprocess = ColumnTransformer([
    ("num", StandardScaler(), Q_num),
    ("cat", OneHotEncoder(handle_unknown="ignore"), Q_cat)
])

attacker = Pipeline([
    ("prep", preprocess),
    ("rf", RandomForestClassifier(n_estimators=100, random_state=0, n_jobs=-1))
])

attacker.fit(X_train, y_train)
pred = attacker.predict(X_test)

print("Attack accuracy:", accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```

```python
results = []

for noise in [0.00, 0.05, 0.10, 0.20, 0.40]:
    anon_tmp = anonymize(df, num_noise=noise, cat_flip_prob=noise/2)

    sample = df.sample(1000, random_state=1)
    Ks = np.array([random_k_for_row(row, anon_tmp) for _, row in sample.iterrows()])

    X = anon_tmp[Q_num + Q_cat]
    y = df[target]

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=0, stratify=y
    )

    attacker.fit(X_train, y_train)
    pred = attacker.predict(X_test)

    results.append({
        "noise": noise,
        "mean_K": Ks.mean(),
        "median_K": np.median(Ks),
        "frac_K_eq_1": np.mean(Ks == 1),
        "frac_K_lt_5": np.mean(Ks < 5),
        "attack_accuracy": accuracy_score(y_test, pred)
    })

results_df = pd.DataFrame(results)
results_df
```
